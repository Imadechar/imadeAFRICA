```python
data = pandas.read_csv("C:\\undata.csv")
data.set_index('Area', inplace=True)
data = data.sort_values('Value')
```

```python
def create_plot(function, label):
    dispersion_list = []
    for year in range(2006, 2015):
        try:
            dispersion_list.append(function(data[data["Period"] == year].Value))
        except ZeroDivisionError:
            dispersion_list.append(0)
    plt.plot(range(2006, 2015), dispersion_list, label=label)
    plt.title('Gross primary graduation ratio')
    plt.ylabel("Value")
    plt.xlabel("Period")
    
create_plot(mean, 'Mean')
create_plot(stdev, 'Standard deviation')
create_plot(median, 'Median')
plt.legend(loc='best')

    
```

```python
create_plot(variance, 'Variance')
plt.legend(loc='best')
```

```python
plt.figure(figsize=(10,10))

def interquartile_range(sample):
    q3, q1 = np.percentile(sample, [75 ,25])
    return q3 - q1

def quartile_range(sample):
    q3, q1 = np.percentile(sample, [75 ,25])
    return (q3 - q1)/2

def _range(sample):
    return sample.max() - sample.min()

def create_plot(fn, label):
    ls = []
    for year in range(2006,2015):
            ls.append(fn(data[data['Period'] == year]['Value']))
              
    plt.plot(range(2006,2015), ls, label=label)
    plt.title('Gross primary graduation ratio')
    plt.ylabel('Value')
    plt.xlabel('Period')
    

#create_plot(statistics.variance, 'Variance')
#create_plot(statistics.stdev, 'Standard Deviation')
create_plot(interquartile_range, 'Interquartile Range')
create_plot(quartile_range, 'Quartile Range')
create_plot(_range, 'Range')

plt.legend(loc='best')
```
